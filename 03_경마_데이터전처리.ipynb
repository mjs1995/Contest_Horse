```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import pandas as pd
```

```python
raw1 = pd.read_csv("서울경마_3년_2_날짜.csv",engine='python')
raw1.head()
```

```python
raw=raw1[["순위","마번","마명","산지","성별","연령","중량","마체중","기수명",
 "조교사명","단승","연승","S-1F","4코너","G-3F","G-1F","날씨","주로습도","등급","거리"]]
raw.head()
```

```python
raw.columns=["순위","마번","마명","산지","성별","연령","중량","마체중","기수명",
 "조교사명","단승","연승","S1F","코너","G3F","G1F","날씨","주로습도","등급","거리"]
```

```python
import datetime
S1F=[]
for i in raw.S1F:
    S1F.append(datetime.datetime.strptime(i,"%M:%S.%f").second+datetime.datetime.strptime(i,"%M:%S.%f").microsecond/1000000)
cor = []
for i in raw.코너:
    cor.append(datetime.datetime.strptime(i,"%M:%S.%f").minute*60+datetime.datetime.strptime(i,"%M:%S.%f").second+datetime.datetime.strptime(i,"%M:%S.%f").microsecond/1000000)
G3F=[]
for i in raw.G3F:
    G3F.append(datetime.datetime.strptime(i,"%M:%S.%f").minute*60+datetime.datetime.strptime(i,"%M:%S.%f").second+datetime.datetime.strptime(i,"%M:%S.%f").microsecond/1000000)
G1F=[]
for i in raw.G1F:
    G1F.append(datetime.datetime.strptime(i,"%M:%S.%f").minute*60+datetime.datetime.strptime(i,"%M:%S.%f").second+datetime.datetime.strptime(i,"%M:%S.%f").microsecond/1000000)
```

```python
S1 = pd.DataFrame(S1F)
co = pd.DataFrame(cor)
G3 = pd.DataFrame(G3F)
G1 = pd.DataFrame(G1F)
```

```python
raw = pd.concat((raw,S1), axis=1)
raw = pd.concat((raw,co), axis=1)
raw = pd.concat((raw,G3), axis=1)
raw = pd.concat((raw,G1), axis=1)
```
